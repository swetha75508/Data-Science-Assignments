# Data Preprocessing

```python
# Handling missing values
df.dropna(inplace = True)  # no missing values are found
```

```python
# remove or detect outlier

columns = df.select_dtypes(include = ['int64']).columns #though all the columns are in numeric i.e numerical_columns = columns
for column in columns:
    Q1 = df[columns].quantile(0.25)
    Q3 = df[columns].quantile(0.75)
    IQR = Q3 - Q1
    lower_extreme = Q1 - 1.5 * IQR
    upper_extreme = Q3 - 1.5 * IQR

    df = df[(df[columns]>=lower_extreme) | (df[columns]<=upper_extreme)]
# '&' or '|' is used to remove outliers
print(df.head())
```

# KMeans Cluster

# choosing number of clusters

WCSS --> Within Clusters Sum of Squares

```python
X = df.drop('Award',axis = 1)
y = df['Award']
```

```python
df.columns
```

```python
# import kmean culster
from sklearn.cluster import KMeans

wcss = []

for i in range(1,11):
    kmeans = KMeans(n_clusters = i, init = 'k-means++', random_state = 200)
    kmeans.fit(df)

    wcss.append(kmeans.inertia_)
# model = kmeans
# n_clusters = no of clusters
# init -> intializing the custer centers
# k-means++ -> it selects the intial cluster centers & speed up the convergence
# random -> initializes the cluster centers by selecting random observations from the dataset.
#kmeans.fit(x) -> fitting the values
# append = adding
```

```python
# plot an elbow graph
plt.figure(figsize= (8,6))
plt.plot(range(1,11),wcss)
plt.scatter(range(1,11),wcss,color = 'red')
plt.title("The Elbow Point Graph")
plt.xlabel("Number of Clusters")
plt.ylabel("WCSS")
plt.show()
```

# Optimum Number of Cluster = 4
Training the K-Means Clustering Model

```python
kmeans = KMeans(n_clusters = 4 , init = 'k-means++', random_state = 200)
# train
labels = kmeans.fit(df)
labels
```

```python
# test
labels = kmeans.fit_predict(df)
labels
```

```python
# labels = KMeans_cluster
df['KMeans_Cluster'] = kmeans.labels_
```

```python
df['KMeans_Cluster'].value_counts()
```

```python
kmeans.cluster_centers_
```

```python
# Aplly PCA method to reduce high dimensional data into two components PCA1 & PCA2 TO make it possible to plot in 2D

# import PCA -->(Principal Component Analysis)
from sklearn.decomposition import PCA

pca = PCA(n_components=2)
pca_data = pca.fit_transform(X,y)


pca_df = pd.DataFrame(data=pca_data, columns=['PC1', 'PC2'])
pca_df.head()

# x=PC1 & y=PC2 : plotting the first two principal component on x & y axes
```

```python
# Visualize K-means clustering
plt.figure(figsize=(10, 8))
sns.scatterplot(x='PC1', y='PC2', hue=kmeans.labels_, data=pca_df, palette='viridis')
plt.scatter(kmeans.cluster_centers_[:,0], kmeans.cluster_centers_[:,1], marker='x', color='red')
plt.title('K-means Clustering')
plt.legend(title='cluster')
plt.show()

# x=PC1 & y=PC2 : plotting the first two principal component on x & y axes
# hue = is a parameter in seaborn scatterplot, used to color the points by cluster labels
```

```python
# import silhouette score
from sklearn.metrics import silhouette_score

# Calculate the Silhouette Score:
kmeans_silhouette = silhouette_score(pca_df,kmeans.labels_)
print(f'K-means Silhouette Score: {kmeans_silhouette}')
```

# Hierarchical Cluster

```python
from scipy.cluster.hierarchy import linkage,dendrogram

linkage_matrix = linkage(pca_df, method = 'ward')
linkage_matrix
```

```python
# plot dendogram
plt.figure(figsize = (10,6))
dendrogram(linkage_matrix)
plt.title("Dendrogram - Hierarchical Clustering")
plt.xlabel("Samples")
plt.ylabel("Distance")
plt.show()
```

```python
# Hierarchical clustering
from sklearn.cluster import AgglomerativeClustering
# train
agglo_cluster = AgglomerativeClustering(n_clusters = 4 , linkage = "ward", metric = "euclidean")
hc = agglo_cluster.fit(df)
hc
```

```python
# test
hc = agglo_cluster.fit_predict(df)
hc
```

```python
# hc = hierarchical cluster
df['Hierarchical_cluster']= agglo_cluster.labels_
# hierarchical clustering doesn't have cluster centers
```

```python
df['Hierarchical_cluster'].value_counts()
```

```python
# Visualize Hierarchical clustering
plt.figure(figsize=(15, 10))
sns.scatterplot(x='PC1', y='PC2', hue=agglo_cluster.labels_, data=pca_df, palette='viridis')
plt.title('Hierarchical Clustering')
plt.legend(title='cluster')
plt.show()
```

```python
# Calculate the silhouette score
agglo_cluster_silhouette = silhouette_score(pca_df, agglo_cluster.labels_)
print(f"Hierarchical Silhouette Score: {agglo_cluster_silhouette}")
```

# DBSCAN Cluster

Density Based Spatial Clustering of Applications with Noise

```python
# import dbscan cluster
from sklearn.cluster import DBSCAN
# train
dbscan_cluster = DBSCAN(eps = 01.7 , min_samples = 10 , metric = 'euclidean')
dbscan = dbscan_cluster.fit(df)
dbscan
```

```python
# TEST
dbscan = dbscan_cluster.fit_predict(df)
dbscan
```

```python
# dbscan = dbscan_clust
df['dbscan_cluster'] = dbscan_cluster.labels_
df.head()
```

```python
df['dbscan_cluster'].value_counts
```

```python

# Visualize dbscan clustering
plt.figure(figsize=(15, 10))
sns.scatterplot(x='PC1', y='PC2', hue=dbscan_cluster.labels_, data=pca_df, palette='viridis',s=100)

# Highlighting noise points (assuming '-1' is the label for noise)
noise_points = pca_df[dbscan_cluster.labels_ == -1]
plt.scatter(noise_points['PC1'], noise_points['PC2'], marker='o', color='red', label='Noise')
plt.title("DBSCA Clustering")
plt.legend(title='cluster')
plt.show()
```
