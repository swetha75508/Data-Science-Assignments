# Association Rule mining

is data mining technique used for the large dataset. It involves finding relationship between variables in the large dataset

```python
transactions = []
for i in range(len(olr)):
    transactions.append(olr.iloc[i, 0].split(','))

olr = pd.DataFrame(transactions).T
```

```python
transactions
```

```python
olr.shape
```

```python
# import libraries for mlxtend
from mlxtend.frequent_patterns import apriori
from mlxtend.frequent_patterns import association_rules

# Transaction encoder will convert dataset into True or False
from mlxtend.preprocessing import TransactionEncoder
```

```python
# transaction encoder
# convert transaction of list to columns using TransactionEncoder
trans_encoder = TransactionEncoder()
trans_ary = trans_encoder.fit_transform(transactions)
# trans_encoder = encoder
# trans_ary = one_hot
# create a new dataframe from trans_ary with the column names
olr1 = pd.DataFrame(trans_ary, columns = trans_encoder.columns_)
olr1
```

```python
trans_ary
```

```python
# convert dataset into 1-0 encoding
def encode_units(x):
    if x== False:
        return 0
    if x== True:
        return 1
```

```python
# .applymap will map the data True to 1 & False to 0(by calling the function encode_units)
olr1 = olr1.applymap(encode_units)
olr1
# olr1 -> dataframe which contains the transaction data
```

```python
# Implement an Apriori algorithm for itemset(data) mining and association rule learning
frequent_itemsets = apriori(olr1, min_support = 0.02, use_colnames = True)
frequent_itemsets
# frequent_itemsets contains all the itemsets which has consider the minimum support threshold = 0.02
# min_support = 0.02 means that it consider the itemsets appeared in at least 2% of transactions.
```

```python
# Apply association rule mining techniques to the pre-processed dataset
#to discver intersting relationship b/w  products that purchased together
rules = association_rules(frequent_itemsets, metric = "confidence", min_threshold = 0.4)
rules

# metric = "confidence" -> sets the metric to be used for generating association rules.
# min_threshold = 0.4-> rule with confidence of atleats 40% will be considered.
```

# Interview Questions

# 1.What is lift and why is it important in Association rules?

Lift is a key metric used ina association rule mining to evaluate the strength and significance of an association.
Lift is used to measure the strength of association b/w to items.
Lift>1 indicates positive association, means A & B are occur together more often than expected by chance.
Lift<1 indicates negative association, means A & B are occur together less often than exoected chance.

# 2.What is support and Confidence. How do you calculate them?

Support: support of an itemset is the proportion of transaction in the dataset.
It measures how frequently an itemset occurs int the dataset.
    
   Support(A) = Number of  transactions containing A)/(Total number of transactions)

# 3.	What are some limitations or challenges of Association rules mining?
Low support:
patterns can be missed if minimum support thresholds are too high even though the items are Rare.

Difficulty in setting Thresholds:
choosing an appropriate support and confidence thresholds is subjective and may contain error.

Interpretation Challenges:
