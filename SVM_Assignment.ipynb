#Data Preprocessing

```python
#	Encode categorical variables
from sklearn.preprocessing import LabelEncoder

labelencoder=LabelEncoder()
for column in df.columns:
    df[column] = labelencoder.fit_transform(df[column])
```

```python
df.head()
```

```python
from sklearn.model_selection import train_test_split

#	Split the dataset into training and testing sets
X = df.drop(['class'], axis=1)
y = df["class"]
# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42, test_size=0.1)
```

#SVM Implementation

```python
from sklearn.svm import SVC

svm = SVC(random_state=42, gamma="auto")
# fit & train the svm model
svm.fit(X_train, y_train)
```

```python
# test the model
y_pred = svm.predict(X_test)
y_pred
```

```python
from sklearn.metrics import accuracy_score, classification_report

#accuracy
print("Test Accuracy: {}%".format(round(svm.score(X_test, y_test)*100, 2)))
# classification report
print("SVM Classifier report: \n\n", classification_report(y_test, y_pred))
```

```python
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay

plt.figure(figsize=(8, 6))
sns.heatmap(confusion_matrix(y_test, y_pred), annot=True, fmt='d', cmap="Blues",
            xticklabels=['Edible', 'Poisonous'], yticklabels=['Edible', 'Poisonous'])
plt.title("Confusion Matrix for SVM Model")
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.show()
```

```python
# experimenting with the different hyperparameter

from sklearn import neighbors, metrics
from sklearn import svm

kernels = ('linear', 'poly', 'rbf', 'sigmoid')

for kernel in kernels:

    # create the model
    model_svm_kern = svm.SVC(kernel=kernel)
    # fit the data(train(train)
    model_svm_kern.fit(X_train, y_train)
    # make predictions(test)
    predictions_kern = model_svm_kern.predict(X_test)
    # accuracy of each kernel
    accuracy_kern = metrics.accuracy_score(y_test, predictions_kern)

    # printing accuracy score for each kernel:
    print("\nKernel {}".format(kernel))
    print("Accuracy = {}\n".format(accuracy_kern))
```

#classification methods with other Models

```python
from sklearn.tree import DecisionTreeClassifier

dt = DecisionTreeClassifier()
dt.fit(X_train, y_train)
```

```python
# test the model
y_pred = dt.predict(X_test)
y_pred
```

```python
# Model Evaluation:

from sklearn.metrics import accuracy_score,confusion_matrix,classification_report

# Accuaracy score
accuracy = accuracy_score(y_test, y_pred)
print(f"Accuracy: {accuracy}")

# classification report
print("\nclassification_report:")
print(classification_report(y_test, y_pred))
```

```python
# Access the feature importances
feature_importances = dt.feature_importances_
feature_importances
```

```python
# create dataframe of dtc
dtc_df = pd.DataFrame({'Feature':X_train.columns, 'Importance': feature_importances})
# sort the Dataframe by importance
dtc_df = dtc_df.sort_values(by = 'Importance', ascending=False)
dtc_df
```

```python
#  Visualize Feature Importances
plt.figure(figsize=(12, 10))
plt.barh(dtc_df['Feature'], dtc_df['Importance'])
plt.xlabel('Importance')
plt.ylabel('Feature')
plt.title('Feature Importances')
plt.gca().invert_yaxis()  # Invert the y-axis to display the highest importance at the top
plt.show()
```
