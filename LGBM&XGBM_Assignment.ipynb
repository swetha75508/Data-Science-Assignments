#Data Preprocessing

```python
df.isnull().sum()
```

```python
# impute Age column with mean of Age
df['Age'].fillna(df['Age'].mean(), inplace = True)
```

```python
#  impute Cabin column with mode
df['Cabin'].fillna(df['Cabin'].mode()[0], inplace = True)
```

```python
# impute Embarked column with mode
df['Embarked'].fillna(df['Embarked'].mode()[0], inplace = True)
```

```python
df.isnull().sum()
```

```python
df['Fare'].unique()
```

```python
Q1=df["Fare"].quantile(0.25)
Q3=df["Fare"].quantile(0.75)


IQR=Q3-Q1  #interqurtile range
lower_bound=Q1-1.5*IQR
upper_bound=Q3+1.5*IQR

# detect outliear
outlier=df[(df["Fare"]<lower_bound)|( df["Fare"] >upper_bound)]
print("Outlier")
print(outlier)


# clean the outliear
df_cleaned_outlier=df[(df["Fare"] >=lower_bound) & (df["Fare"] <= upper_bound)]

df_cleaned_outlier
```

```python
df = df_cleaned_outlier.copy()
df
```

```python
df.drop('Name', axis = 1, inplace = True)
```

```python
# label encoding
from sklearn.preprocessing import LabelEncoder

for col in df.columns:
    le=LabelEncoder()
    df[col]=le.fit_transform(df[col])
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
std_scaled_df = scaler.fit_transform(df)
std_scaled_df
```

```python
std_scaled_df = pd.DataFrame(std_scaled_df, columns=df.columns)
std_scaled_df.head()
```

#Buiding Predictive model

```python
X = df.drop('Survived', axis= 1)
y = df['Survived']


from sklearn.model_selection import train_test_split
#	Split the preprocessed dataset into training and testing sets.
X_train,X_test, y_train,y_test = train_test_split(X,y, test_size = 0.2, random_state = 30)
```

```python
X_train
```

```python
X_test
```

```python
y_train
```

```python
y_test
```

```python
from xgboost import XGBClassifier
from lightgbm import LGBMClassifier
from sklearn.metrics import accuracy_score,classification_report

# xgb model building and evaluation
xgb_model = XGBClassifier()
xgb_model.fit(X_train, y_train)
y_pred_xgb = xgb_model.predict(X_test)
xgb_accuracy = accuracy_score(y_test, y_pred_xgb)
xgb_classification_report = (y_test, y_pred_xgb)
print(f"XGB - Accuracy: {xgb_accuracy}")

```

```python
# lgb model building and evaluation
lgb_model = LGBMClassifier()
lgb_model.fit(X_train, y_train)
y_pred_lgb = lgb_model.predict(X_test)
lgb_accuracy = accuracy_score(y_test, y_pred_lgb)
lgb_classification_report = (y_test, y_pred_xgb)
print(f"LGB - Accuracy: {lgb_accuracy}")
```

# Comparative Analysis

```python
print(f"The accuracy of the XGB is {xgb_accuracy}")
print(f" The accuracy of the LGB is {lgb_accuracy}")
```

```python
xgb_report = classification_report(y_test, y_pred_xgb)
lgb_report = classification_report(y_test, y_pred_lgb)

print("XGB Classification Report:")
print(xgb_report)

print("LGB Classification Report:")
print(lgb_report)
```

```python
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay

cm = confusion_matrix(y_test, y_pred_xgb)
cm_xgb = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels = ['Not Survived','Survived'])
cm_xgb.plot()

cm = confusion_matrix(y_test, y_pred_lgb)
cm_lgb = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels= ['Not Survived','Survived'])
cm_lgb.plot()
```
